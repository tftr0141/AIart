## オプショナル（パスワード類生成）

````python
#@markdown オプショナル

#@markdown 強力なパスワード生成セル。人間が考えるパスワードよりもランダムで長いパスワードの方が安全。あと考えるの面倒だし、Colabから出せるようにしておけば便利だ<br>
#@markdown ASCII全文字を生成対象にした方が安全なんだけど、英数字だけで64桁なら十分総当たり耐性あるから大丈夫でしょ<br><br>
#@markdown デフォルトでは64桁のパスワード

#@markdown パスワードに偏りがあっても気にしなくてOK。さすがに16連続Aとかなら再生成した方が良いが。寧ろ偏りがなかったり、似たような文字を省いたパスワードの方が暗号強度が低い。後パスワード文字数が分かると危なくない？って思うかもしれないけど、ランダム生成64桁クラスなら気にしなくていい<br>
#@markdown 他の人と共同でこのノート使ってたり、脆弱性発覚とかなら話は別だけど、パスワードは毎セッション変える必要はない。世間ではパスワードは定期的に変更しましょう！とか言われているが、定期的に変更する場合は変更が面倒で、人間が脆弱なパスワードを使いがち（Colab_20221120とか）なので、充分強力なランダム生成ならその必要はない<br><br>
#@markdown ただし、マシンスペックの向上で危殆化する場合があるので、そこら辺は柔軟に。何にせよ64桁が危殆化するのには100年以上は必要になるとは思う
#@markdown 総当たりなら1兆年くらいかければ多分突破できるんじゃない？
import secrets
import string

def get_password(length: int) -> str:
    # secretsはrandomとかと違って暗号学的に安全な乱数を生成できる。
    pass_chars = string.ascii_letters + string.digits
    password = ''.join(secrets.choice(pass_chars) for i in range(length))
    return password

for i in range(10):
  # ここに桁数。個人的には8~16とかの貧弱なパスワードは使わない方が良い。サイトが受け付けてるなら256字とかでもいいくらい
    print(get_password(64)) 

# Colabのセッション中にずっと残るの気持ち悪いんでアンインポート
del secrets
del string
del get_password
````

# 全共通セットアップ
必ず実行

````python
def kill_weak_password(password: str,user_mode: bool=False) -> None:
    password = str(password)

    if " " in password:
        raise ValueError("ユーザー名かパスワードに半角スペースが含まれています")

    if "\n" in password or "\r" in password:
        raise ValueError("ユーザー名かパスワードにCRもしくはLF改行コードが含まれています")

    if "password" == password:
        raise ValueError("多分サンプルのまま使ってない？パスワードかユーザー名がpasswordになってる")

    if "qwerty" == password:
        raise ValueError("多分サンプルのまま使ってない？パスワードかユーザー名がqwertyになってる")

    if len(password) >= 65536:
        raise ValueError("さすがに65536字以上のパスワードかユーザー名はちょっとね")

    if not password.isascii():
        raise ValueError("全角とか日本語とか含まれてない？めんどくさそうなんで、ASCII以外は弾くよ")

    if user_mode:
        return

# フールプルーフの観点からこっちで弱いパスワードは止める
    if len(password) < 32:
        print("32字未満のパスワードは使用不可。パスワードマネージャーとかの仕様でどうしてもそうぜざるを得ないならコード直接弄って無力化して")
        raise ValueError("31字以下の脆弱なパスワードではこのノートは動作しません。このエラーが出たからって後ろに大量に1つけて通るようにするとかはNG")

print("finish")
````

### オプショナル シンボリックリンクのアンリンク
Google Driveに丸ごとコピーする前に実行

WEBUI止めて rm系のファイル・ディレクトリ削除コマンド打つときはこのセルを実行した方が安全<br><br>
シンボリックリンクを”比較的”安全な方法で除去する


```
unlinkコマンドはrm -rfより安全ってだけで絶対に安全って訳でもないから注意してな！
```


言うまでもなくアンリンクしたら、WEBUIは次からColabマシン上のモデルファイルとか拡張機能探し始める。また動かすなら再リンクを実行

````python
#@markdown SDPathとかは”初期セットアップ2 シンボリックリンクの生成”と同一にしないとダメ<br>
#@markdown Stopはこのセル実行するときに再生みたいなの押したでしょ？それ
project_dir = "/content/stable-diffusion-webui/"
sd_path = "{project_dir}models/Stable-diffusion"
extension_path = "{project_dir}extensions"

%cd "{project_dir}"

!ls -la | grep -- "->"
!echo "各種シンボリックリンク先表示してるけど、リンク名->リンク先に全部なってる？もしなってなかったらすぐにStopボタン押下。なってるならplease key:のすぐ右をクリックしてEnter: "
!echo "何も表示されてない場合も即Stop"
!read -p "please key:"

%cd "{project_dir}"
!unlink models
!unlink extensions


# 後片付け
l = [ project_dir, sd_path, extension_path]
for i in l:
    i = None  #pythonならこれで参照切れるはず。Javaだと参照の値渡しとか分けわからん参照なんで、切れないけど
l = None
````
